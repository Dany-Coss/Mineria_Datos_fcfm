```python
moda_overall = df["overall"].mode
print(moda_overall)
varianza_edad = df["age"].var(ddof=0)
print("\nLa varianza en la edad es: ",varianza_edad)
desviacion_edad = df["age"].std(ddof=0)
print("\nLa desviación estandar en la edad es: ",desviacion_edad)

varianza_overall = df["overall"].var(ddof=0)
print("\nLa varianza del nievel de portero es: ",varianza_overall)
desviacion_overall = df["age"].std(ddof=0)
print("\nLa desviación estandar en el nivel es: ",desviacion_overall)

varianza_valor = df["value_eur"].var(ddof=0)
print("\nLa varianza del valor en euros de portero es: ",varianza_valor)
desviacion_valor = df["value_eur"].std(ddof=0)
print("\nLa desviación estandar del valor en euros del portero es: ",desviacion_valor)
```

```python
import matplotlib.pyplot as plt

x = df["overall"]
y = df["value_eur"]

plt.plot(x,y)
plt.xlabel("Nivel del portero")
plt.ylabel("Precio del portero (millones de euros)")
plt.title("Comportamiento del precio con releación al nivel")
plt.show()

x1 = df["overall"]
y1 = df["wage_eur"]

plt.plot(x1,y1)
plt.xlabel("Nivel del portero")
plt.ylabel("Precio del portero (millones de euros)")
plt.title("Comportamiento del precio con releación al nivel")
plt.show()

x2 = df["age"]

plt.hist(x2)
plt.xlabel("Edad del portero")
plt.title("Potencial con relación a la edad")
plt.show()



#https://numython.github.io/posts/2016/02/graficar-en-python-con-matplotlib-y/
```
